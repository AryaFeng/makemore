```python
words = open('names.txt','r').read().splitlines()
words[:10]
```

```python
b = {}  # 用来统计二元组出现的频次
for w in words:
    # 添加起始和结束表示
    chs = ['<S>'] + list(w) + ['<E>']
    for c1,c2 in zip(chs,chs[1:]):
        bigram = (c1,c2)
        b[bigram] = b.get(bigram,0) + 1   # 如果是第一次是0，后续出现一次加1
        # print(c1,c2)
```

```python
sorted(b.items(), key=lambda kv:-kv[1])
```

```python
import torch
```

```python
a = torch.zeros((26+2,28),dtype=torch.int32)
# 然后将上面的b 行和列分别都是每个char，值代表他们的个数
for k,v in b.items():
    # 处理特殊字符
    if k[0] == '<S>':
        a[0,ord(k[1])-ord('a')] = v
    elif k[1] == '<E>':
        a[ord(k[0])-ord('a'),27] = v

a
```

```python
chars = list(set("".join(words)))
stoi = {s:i for i,s in enumerate(chars)}
stoi['<S>'] = 26
stoi['<E>'] = 27
stoi
```
